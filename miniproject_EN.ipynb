# Task 5 "Let's start exploring the data! Which countries have the highest number of successful bookings? A booking is considered successful if it was not canceled (variable `is_canceled`). Select the top 5 countries as your answer."


```python
bookings.query("is_canceled == 1") \
    .groupby(['country'], as_index=False) \
    .aggregate({'is_canceled': 'count'}) \
    .sort_values('is_canceled', ascending=False).head(5)
# The task requires finding successful hotel bookings by country (i.e., non-canceled bookings) using the query filter.
# Group by country.
# Aggregate by the count of successful bookings (use `count()` method).
# Then, sort using `sort_values()`.
# The `head()` function provides the top entries.
```

```python
bookings['country'].value_counts().sort_values(ascending=False)
```

# Task 8 "Now analyze the scheduled arrival dates (`arrival_date_year`)."  
## In which month were bookings most frequently made in 2016? Did the most popular month change in 2017?

```python
# Filter rows in the DataFrame `bookings`, selecting only those where the column `is_canceled` has the value 1 (canceled bookings).
# Group by arrival year and month (`arrival_date_year`, `arrival_date_month`).
# Count the number of canceled bookings for each group.
# Sort groups in descending order by the number of canceled bookings.
bookings.query("is_canceled == 1") \
    .groupby(['arrival_date_year', 'arrival_date_month'], as_index=False) \
    .aggregate({'is_canceled': 'count'}) \
    .sort_values('is_canceled', ascending=False)
```

# Task 9 "Group the data by year, then check in which month (`arrival_date_month`) City Hotel bookings were most frequently canceled in 2015, 2016, and 2017."

```python
# Filter rows in the DataFrame `bookings`, selecting only those where `is_canceled` is 1 (canceled bookings).*
# Group by `hotel`, `arrival_date_month`, and `arrival_date_year`.
# Grouping is done by hotel type (e.g., city or resort), month, and arrival year.
# Count the number of canceled bookings for each group.
# Sort groups by the number of canceled bookings in descending order.

bookings.query("is_canceled == 1") \
    .groupby(['hotel', 'arrival_date_month', 'arrival_date_year'], as_index=False) \
    .aggregate({'is_canceled': 'count'}) \
    .sort_values('is_canceled', ascending=False)
```

# Task 12 "Not all bookings were completed successfully (`is_canceled`), so we can calculate the number of lost clients. In other words, we can calculate a metric called Churn Rate."

## Churn rate is the percentage of subscribers (e.g., to site push notifications) who unsubscribed from communication or stopped using the service within a specific period. In other words, it’s the ratio of the number of users who left to the total number of users, expressed as a percentage.

## In our case, Churn Rate is the percentage of clients who canceled their booking. Let's see how this metric is related to whether clients have children.

## - Create a column `has_kids` that is `True` if a client indicated at least one child (`total_kids` > 0), otherwise `False`. Then, check which user group has the higher churn rate.

## *For the answer, provide the highest churn rate as a percentage rounded to 2 decimal places (e.g., 0.24563 becomes 24.56%).*

```python
bookings['has_kids'] = bookings['total_kids'].apply(lambda x: True if x >= 1 else False)
```

```python
bookings.query('is_canceled == 1') \
    
```

```python
# Calculate the number of canceled bookings with children.
canceled_with_kids = bookings.query('is_canceled == 1 and has_kids == True') \
    .groupby('has_kids', as_index = False) \
    .aggregate({'is_canceled': 'count'})
canceled_with_kids
```

```python
# Calculate the number of canceled bookings without children.
canceled_with_no_kids = bookings.query('is_canceled == 1 and has_kids == False') \
    .groupby('has_kids', as_index = False) \
    .aggregate({'is_canceled': 'count'})
canceled_with_no_kids
```

```python
# Calculate the number of non-canceled bookings with children.
no_canceled_with_kids = bookings.query('is_canceled == 0 and has_kids == True') \
    .groupby('has_kids', as_index= False) \
    .aggregate({'is_canceled': 'count'})
no_canceled_with_kids
```

```python
# Calculate the number of non-canceled bookings without children.
no_canceled_with_no_kids = bookings.query('is_canceled == 0 and has_kids == False') \
    .groupby('has_kids', as_index= False) \
    .aggregate({'is_canceled': 'count'})
no_canceled_with_no_kids
```

```python
# Also, calculate the difference in the percentage of non-canceled bookings with children vs. canceled bookings with children.
((canceled_with_kids.is_canceled / no_canceled_with_kids.is_canceled) * 100).round(2)
```

```python
# Calculate the difference in non-canceled bookings without children vs. canceled bookings without children.
((canceled_with_no_kids.is_canceled / no_canceled_with_no_kids.is_canceled) * 100).round(2)
```

```python
# Calculate the total number of bookings with children.
total_bookes_has_kids = bookings.query('has_kids == True') \
    .groupby('has_kids', as_index= False) \
    .aggregate({'is_canceled': 'count'})
total_bookes_has_kids
```

```python
# Calculate the difference in canceled bookings with children vs. the total number of bookings with children.
((total_canceled_bookes_has_kids.is_canceled / total_bookes_has_kids.is_canceled) * 100).round(2)
```

```python
# Calculate the total number of bookings without children.
total_bookes_has_no_kids = bookings.query('has_kids == False') \
    .groupby('has_kids', as_index= False) \
    .aggregate({'is_canceled': 'count'})
total_bookes_has_no_kids
```
