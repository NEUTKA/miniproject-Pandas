# Задача 5 "Перейдем к исследованию данных! Пользователи из каких стран совершили наибольшее число успешных бронирований? Бронирование считается успешным, если в дальнейшем не было отменено (переменная is_canceled). В качестве ответа выберите страны, входящие в топ-5."

```python
bookings.query("is_canceled == 1") \
    .groupby(['country'], as_index=False) \
    .aggregate({'is_canceled': 'count'}) \
    .sort_values('is_canceled', ascending=False)
# В задании требовалось найти успешное бронирование отелей по странам(т.е. не отмененных бронирований) query - это фильтр.
# затем группируем по странам
# аггрегируем по количеству успешных бронирований (считаем по методу count())
# затем сортируем с помощью метода sort_values() 
# функция head() дает количество первых колонок
```

# Задача 8 "Теперь проанализируйте даты запланированного прибытия (arrival_date_year)."
# На какой месяц чаще всего оформляли бронь в 2016 году? Изменился ли самый популярный месяц в 2017?

```python
# Фильтруем строки в DataFrame `bookings`, выбирая только те, где колонка 'is_canceled' имеет значение 1 (отмененные бронирования).
    # Группируем данные по годам и месяцам прибытия ('arrival_date_year', 'arrival_date_month').
    # Для каждой группы считаем количество отмененных бронирований.
    # Сортируем группы по убыванию количества отмененных бронирований.
bookings.query("is_canceled == 1") \
    .groupby(['arrival_date_year', 'arrival_date_month'], as_index=False) \
    .aggregate({'is_canceled': 'count'}) \
    .sort_values('is_canceled', ascending=False)
```

# Задача 9 "Сгруппируйте данные по годам, а затем проверьте, на какой месяц (arrival_date_month) бронирования отеля типа City Hotel отменялись чаще всего в 2015? 2016? 2017?" 

```python
# Фильтруем строки в DataFrame `bookings`, выбирая только те, где колонка 'is_canceled' имеет значение 1 (отмененные бронирования).
   # Группируем данные по категориям 'hotel', 'arrival_date_month', и 'arrival_date_year'.
    # Группировка осуществляется по типу отеля (например, городской или курортный), месяцу и году прибытия.
       # Для каждой группы считаем количество отмененных бронирований.
    # Сортируем группы по количеству отмененных бронирований в порядке убывания.
bookings.query("is_canceled == 1") \
    .groupby(['hotel', 'arrival_date_month', 'arrival_date_year'], as_index=False) \
    .aggregate({'is_canceled': 'count'}) \
    .sort_values('is_canceled', ascending=False)
```

# Задача 10 "Посмотрите на числовые характеристики трёх колонок: adults, children и babies. Какая из них имеет наибольшее среднее значение?"

```python
# Группируем данные в DataFrame `bookings` по комбинациям значений в колонках 'adults', 'children' и 'babies'.
bookings.groupby(['adults', 'children', 'babies'], as_index=False) \
    .aggregate({'adults': 'mean', 'children': 'mean', 'babies': 'mean'}) \
    .sort_values(by=['adults', 'children', 'babies'], ascending = [False, False, False]) 
    # Выполняем агрегацию для каждой группы, считая средние значения по колонкам 'adults', 'children' и 'babies'.
    # Сортируем полученные данные по убыванию значений 'adults', затем 'children' и 'babies'.
```

# Задача 12 "Не все бронирования завершились успешно (is_canceled), поэтому можно посчитать, сколько клиентов было потеряно в процессе. Иными словами, посчитать метрику под названием Churn Rate."

<font size="4"> Churn rate (отток, коэффициент оттока) – это процент подписчиков (например, на push-уведомления от сайта), которые отписались от канала коммуникации, отказались от услуг сервиса в течение определенного периода времени. Иными словами, представляет собой отношение количества ушедших пользователей к общему количеству пользователей, выраженное в процентах.</font><br>

<font size="4"> В нашем случае Churn Rate - это процент клиентов, которые отменили бронирование. Давайте посмотрим, как эта метрика связана с наличием детей у клиентов!</font><br>

<font size="4"> Создайте колонку has_kids, которая принимает значение True, если клиент при бронировании указал хотя бы одного ребенка (total_kids), в противном случае – False. Далее проверьте, среди какой группы пользователей показатель оттока выше. </fomt><br>

В качестве ответа укажите наибольший % оттока, округленный до 2 знаков после точки (то есть доля 0.24563 будет 24.56% и в ответ пойдёт 24.56)

```python
bookings['has_kids'] = bookings['total_kids'].apply(lambda x: True if x >= 1 else False)
```

```python
bookings.query('is_canceled == 1') \
    
```

```python
# посчитаем количество отмененных бронирований с наличием детей у юзеров.
canceled_with_kids = bookings.query('is_canceled == 1 and has_kids == True') \
    .groupby('has_kids', as_index = False) \
    .aggregate({'is_canceled': 'count'})
canceled_with_kids
```

```python
# посчитаем количество отмененных бронирований у юзеров у которых нет детей.
canceled_with_no_kids = bookings.query('is_canceled == 1 and has_kids == False') \
    .groupby('has_kids', as_index = False) \
    .aggregate({'is_canceled': 'count'})
canceled_with_no_kids
```

```python
# посчитаем количество не отмененных бронирований с детьми.
no_canceled_with_kids = bookings.query('is_canceled == 0 and has_kids == True') \
    .groupby('has_kids', as_index= False) \
    .aggregate({'is_canceled': 'count'})
no_canceled_with_kids
```

```python
# Посчитаем количество не отмененных бронирований без детей.
no_canceled_with_no_kids = bookings.query('is_canceled == 0 and has_kids == False') \
    .groupby('has_kids', as_index= False) \
    .aggregate({'is_canceled': 'count'})
no_canceled_with_no_kids
```

```python
# Так же для интереса посчитаем на сколько процент не отмененных бронирований с детьми больше чем отмененных с детьми.
((canceled_with_kids.is_canceled / no_canceled_with_kids.is_canceled) * 100).round(2)
```

```python
# Посчитаем насколько неотмененных бронирований без детей больше чем отмененных без детей.
((canceled_with_no_kids.is_canceled / no_canceled_with_no_kids.is_canceled) * 100).round(2)
```

```python
# Теперь посчитаем общее количество бронирований с детьми
total_bookes_has_kids = bookings.query('has_kids == True') \
    .groupby('has_kids', as_index= False) \
    .aggregate({'is_canceled': 'count'})
total_bookes_has_kids
```

```python
# Общее количество отмененных бронирований с детьми
total_canceled_bookes_has_kids = bookings.query('is_canceled == 1 and has_kids == True') \
    .groupby('has_kids', as_index = False) \
    .aggregate({'is_canceled': 'count'})
total_canceled_bookes_has_kids
```

```python
# Теперь посчитаем насколько количество отмененных бронирований с детьми выше чем общее количество бронирований с детьми.
((total_canceled_bookes_has_kids.is_canceled / total_bookes_has_kids.is_canceled) * 100).round(2)
```

```python
# общее количество бронирований без детей
total_bookes_has_no_kids = bookings.query('has_kids == False') \
    .groupby('has_kids', as_index= False) \
    .aggregate({'is_canceled': 'count'})
total_bookes_has_no_kids
```

```python
# Общее количество отмененных бронирований без детей
total_canceled_bookes_has_no_kids = bookings.query('is_canceled == 1 and has_kids == False') \
    .groupby('has_kids', as_index= False) \
    .aggregate({'is_canceled': 'count'})
total_canceled_bookes_has_no_kids
```

```python
# Теперь посчитаем насколько количество отмененных бронирований без детей выше чем общее количество бронирований без детей.
((total_canceled_bookes_has_no_kids.is_canceled / total_bookes_has_no_kids.is_canceled) * 100).round(2)
```
